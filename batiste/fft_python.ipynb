```python
import numpy as np

def fft_python(x,N):
    
    if(N==1):
        return x
    x_even = x[0::2]
    x_odd = x[1::2]

    y0 = fft_python(x_even,N/2)
    y1 = fft_python(x_odd,N/2)

    yk = np.zeros(int(N),dtype=complex)

    for i in range(0,int(N/2)):
        omega = np.exp(-2j*np.pi*i/N)
        yk[i] = y0[i] + y1[i]* omega
        yk[int(i + N/2)] = y0[i] - y1[i]*omega
    
    return yk
```

```python
from scipy.io import wavfile
import datetime
samplerate, data = wavfile.read('son/Ultrasonic_-_Slap_House_Essentials_-_Clap_04.wav')

import matplotlib.pyplot as plt
plt.plot(data[:,1])
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.show()
```

```python
system_time = datetime.datetime.now()
y = fft_python(data[0:pow(2,15),1],pow(2,15))
system_time2 = datetime.datetime.now()
```
